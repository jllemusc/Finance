The first step is activate all the libraries neccesaries  

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import yfinance as yf
```

**1)** Download the latest daily close BTC-to-USD data, from 2021-01-01 up to this day

Instead of download the data from a csv document using a package called yfinance, we can download the data directly to our Jupyter Notebook. We need to be carefull to wraggling the right interval of data, because using 2021-01-01 we take information from the last day of 2020

```python
#WE download the information from Yahoo
BTC = yf.Ticker("BTC-USD")
#We take from the data set the column 'Close'
rates = BTC.history(start='2021-01-02')['Close']
#We show the first 5 lines to ensure we are working with the right data
rates.head(5)
```

**3)** Call matplotlib to draw the data

It always looks good when the data is in a plot. In this way, we can visualise some patrons from the data.

We decide to change the plot size based on the predetermined values from matplotlib.Therefore, we can notice more details in the plot.

```python
plt.figure(figsize=(10,5))
plt.plot(rates, linestyle='dashed')
plt.title('BTC to USD')
plt.xlabel('Date')
plt.ylabel('Dolars')
```

**4)** For the data between '2021-01-01' to the '2021-12-31' 365 days, calculate arithmetic mean median minimum maximum standard deviation and interquartile range.

```python
rates_2021 = rates[0:365]
print('The data start in %s  and finish in %s'  % (rates_2021.index[0], rates_2021.index[-1]))

# Arithmetic mean
print('the arithmetic mean price of Bitcoin in 2021 on yahoo Finance was: $%.2f'
      % (rates_2021.mean()))

# Median
print('the median price of Bitcoin in 2021 on yahoo Finance was: $%.2f' 
      % (rates_2021.median()))

# Standard deviation
print('the standard deviation of Bitcoin in 2021 on yahoo Finance was: $%.2f'
      % (rates_2021.std()))

# maximum price
print('the maximum price of Bitcoin in 2021 on yahoo Finance was: $%.2f' 
      % (rates_2021.max()))

# minimum price
print('the minimum price of Bitcoin in 2021 on yahoo Finance was: $%.2f' 
      % (rates_2021.min()))

# First quartile (Q1)
Q1 = np.percentile(rates_2021, 25, interpolation = 'midpoint')
  
# Third quartile (Q3)
Q3 = np.percentile(rates_2021, 75, interpolation = 'midpoint')
  
# Interquaritle range (IQR)
IQR = Q3 - Q1
  
print('the interquantile range price of Bitcoin in 2021 on yahoo Finance was: $%.2f'
      % (IQR))
```

**5)** Determine the  day numbers with the lowest and highest observations.
In this case we use the number of datapoints locate in the 5% low and the top 5%.

```python
#we identify the lower band
under5 = np.percentile(rates_2021, 5, interpolation = 'midpoint')
print ('The lowerband with the better prices to find a cheap btc are those under %.2f' %under5)
#we identify the upper band
over5 = np.percentile(rates_2021, 95, interpolation = 'midpoint')
print ('The the higher band with the better prices to sell btc are those over %.2f' %over5)
```

```python
#we create an empty dataframe to put the information relative to the upperband
filterup = []

#we create a for loop to identify the elements 
for element in rates_2021:
    if element > over5:
        filterup.append(True)
    else:
        filterup.append(False)
        
upper = rates_2021[filterup]

#We print the 5% better days to sell BTC
print(upper)
```

```python
#we create an empty dataframe to put the information relative to the lowerband
filterdown = []

#we create a for loop to identify the elements 
for element in rates_2021:
    if element < under5:
        filterdown.append(True)
    else:
        filterdown.append(False)
down = rates_2021[filterdown]

#We print the 5% better days to buy BTC

print(down)
   
```

**6)** Use and horizontal box-and-whisker plot

```python
# First quartile (Q1)
Q1 = np.percentile(np.diff(rates_2021), 25, interpolation = 'midpoint')
# Third quartile (Q3)
Q3 = np.percentile(np.diff(rates_2021), 75, interpolation = 'midpoint')
# Interquaritle range (IQR)
print(Q1)
print(Q3)
IQR = Q3 - Q1
print(IQR)
```

```python
#We change the size of the plot in order to notice more details
plt.figure(figsize=(10,2.5))
#we add some extra characteristics to the line to obtain the plot under the requirements
plt.boxplot(np.diff(rates_2021),vert=False)
#It was added a title
plt.title('Daily change price of BTC')
```

The daily change in price in the BTC boxplot indicates how much the price could change from one day to another. We can confirm that 25% of the time in 2021, the daily positive change in bitcoin was 1138, and the other 25 % of the time, the change was around $ 1011.It means that for 50 % of the days in 2021, we would expect the BTC price would be increasing or decreasing by 1,000. The low whisker and the Q1 have an equivalence of around 2900, and the Q3 and the high whisker have an equivalence of around 2900. The most relevant conclusion in this plot is the outliers. We only have 4 positive outliers vs 8 negative outliers. We can conclude and expect that the BTC price trend to have more strong drops from one day to another than positive extreme changes.  

## Extra features

```python
plt.figure(figsize=(10,2.5))
plt.boxplot(np.diff(rates_2021),vert=False)
plt.boxplot(np.diff(rates_2021),vert=False, showmeans = True)
plt.title('Daily change price of BTC')
```

**2)** Count (programmatically) how many outliers are there on the boxplot 

```python
#calculate the number of negative outliers. 
print('The number of outliers in the daily change price of Bitcoin when it drops is: %.f'
      % ((np.diff(rates)) < (Q1 - 1.5 * IQR)).sum())
#calculate the number of positive outliers. 
print('The number of outliers in the daily change price of Bitcoin when it rises is: %.f'
      % ((np.diff(rates)) > (Q3 + 1.5 * IQR)).sum())
```

As we explained before, the number of outliers is concentrated in the negative area, meaning that even though the price of bitcoin rose in 2021, it was most likely a really hard drop on any day (relation 1 to 2) instead of a suddenly high spike in the price of this cryptocurrency.
